```python
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
from keras.layers.normalization import BatchNormalization
from keras.regularizers import l2
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import keras
```

```python
train_data_path = r'C:\Users\Sameer\OneDrive\Desktop\Horse Human Prediction\train'
validation_data_path = r'C:\Users\Sameer\OneDrive\Desktop\Horse Human Prediction\validation'
```

```python
print(f'Training data path:-\n{train_data_path}')
print(f'Validation data path:-\n{train_data_path}')
```

```python
#Data preprocessing
#For train data
print('Training Data Preprocessing:-')
training_data_gen = ImageDataGenerator(rescale=1./255, rotation_range=40, width_shift_range=0.2, height_shift_range=0.2, shear_range=0.2, zoom_range=0.2, horizontal_flip=True, fill_mode='nearest')
train_data = training_data_gen.flow_from_directory(train_data_path, target_size=(150, 150), batch_size=32, class_mode='binary')
print(f'Classes:-\n{train_data.class_indices}')
#For validation data
print('\nValidation Data Preprocessing:-')
validation_data_gen = ImageDataGenerator(rescale=1./255)
valid_data = validation_data_gen.flow_from_directory(validation_data_path, target_size=(150, 150), batch_size=32, class_mode='binary')
print(f'Classes:-\n{valid_data.class_indices}')
```

```python
#For visulazing images
def plotImages(images_arr):
    fig, axis = plt.subplots(1, 5, figsize=(20, 20))
    axis = axis.flatten()
    for img, ax in zip(images_arr, axis):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
print('Train Data:-')
train_images = [train_data[0][0][0] for i in range(5)]
plotImages(train_images)
print('Validation Data:-')
validate_images = [valid_data[0][0][0] for i in range(5)]
plotImages(validate_images)
```

```python
#Build a conviotunial neural network
cnn_model = keras.models.Sequential([
                                    keras.layers.Conv2D(filters=32, kernel_size=7, input_shape=[150, 150, 3], kernel_regularizer=l2(l=0.01)),
                                    BatchNormalization(),
                                    keras.layers.MaxPooling2D(pool_size=(2,2)),
    
                                    keras.layers.Conv2D(filters=64, kernel_size=5),
                                    BatchNormalization(),
                                    keras.layers.MaxPooling2D(pool_size=(2,2)),
    
                                    keras.layers.Conv2D(filters=128, kernel_size=3),
                                    BatchNormalization(),
                                    keras.layers.MaxPooling2D(pool_size=(2,2)),
    
                                    keras.layers.Conv2D(filters=256, kernel_size=3),
                                    BatchNormalization(),
                                    keras.layers.MaxPooling2D(pool_size=(2,2)),
    
                                    #Build a neural netword
                                    keras.layers.Flatten(), 
                                    #Input layer
                                    keras.layers.Dense(units=128, activation='relu'),
                                    BatchNormalization(),
                                    keras.layers.Dropout(0.5),
                                    keras.layers.Dense(units=256, activation='relu'),
                                    BatchNormalization(),
                                    keras.layers.Dropout(0.5),
                                    #Output layer
                                    keras.layers.Dense(units=2, activation='softmax')
])
```

```python
#Compile the model
cnn_model.compile(optimizer=Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Save model
model_path = 'C:/Users/Sameer/OneDrive/Desktop/Horse Human Prediction/model.h5'
checkpoint = ModelCheckpoint(model_path, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
```

```python
#Prediction
def prediction(model, images):
    test_image = image.load_img(images, target_size=(150,150))
    test_image = image.img_to_array(test_image)/255
    test_image = np.expand_dims(test_image, axis=0)
    
    ans = model.predict(test_image).round(3)
    pred = np.argmax(ans)
    
    print(f'{ans}----->>>>>{pred}')
    
    if pred == 0:
        print('Horse')
    else:
        print('Human')
```

```python
#For model 1
for images in [horse_1, horse_2, horse_3, horse_4, horse_5, human_1, human_2, human_3, human_4, human_5]:
    prediction(model_1, images)
```

```python
#For model 2
for images in [horse_1, horse_2, horse_3, horse_4, horse_5, human_1, human_2, human_3, human_4, human_5]:
    prediction(model_2, images)
```
